```python
import math

class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
    def calculate_volume(self):
        return math.pi * self.radius**2 * self.height
    def calculate_surface_area(self):
        return 2 * math.pi * self.radius * (self.radius + self.height)
    def __str__(self):
        """Возвращает строковое представление объекта."""
        return f"Цилиндр:\nРадиус: {self.radius}\nВысота: {self.height}"
    def show_info(self):
       """Выводит информацию об объекте с вычислениями"""
       volume = self.calculate_volume()
       surface_area = self.calculate_surface_area()
       print(self)
       print(f"Объем: {volume:.2f}")
       print(f"Площадь поверхности: {surface_area:.2f}\n")

```
